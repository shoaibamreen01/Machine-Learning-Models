# **Deep learning**
**Deep learning** is based on artificial neural networks, inspired by how the human brain works. These networks have multiple layers (hence “deep”), allowing them to automatically learn complex features from raw data.
How It Works

A deep learning model consists of:

**Input layer** – receives data (images, text, etc.)

**Hidden layers** – process and extract features

**Output layer** – produces prediction

**1. Weight**
A weight tells the neural network how important an input is.

It controls the influence of input values.

**2. Bias**
A bias helps the neural network adjust the output.

It allows the model to shift the prediction up or down.



**Difference Between Machine Learning and Deep Learning**

| Machine Learning          | Deep Learning                |
| ------------------------- | ---------------------------- |
| Small/medium data         | Large data                   |
| Manual feature extraction | Automatic feature extraction |
| Faster training           | Slower training              |
| Simple models             | Neural networks              |


**Neural Network Learning Process**

Input → Weight → Bias → Output → Error → Update

```python
import numpy as np

# Input data
X = np.array([1, 2, 3, 4])

# Real output
y = np.array([2, 4, 6, 8])

```

```python
# Initial weight and bias
weight = 0.5
bias = 0.0
```

```python
# Learning rate
lr = 0.01
```

```python

# Training
for epoch in range(1000):

    # Prediction
    y_pred = weight * X + bias

    # Error
    error = y - y_pred

    # Update weight and bias
    weight = weight + lr * np.mean(error * X)
    bias = bias + lr * np.mean(error)
```

```python
# Final prediction
print("Learned Weight:", round(weight, 2))
print("Learned Bias:", round(bias, 2))
```

```python
# Test
test = 5
prediction = weight * test + bias

print(f"Prediction for {test} is:", round(prediction, 2))
```
